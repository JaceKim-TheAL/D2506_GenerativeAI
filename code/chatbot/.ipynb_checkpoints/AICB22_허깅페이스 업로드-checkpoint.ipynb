### 1. Í∏∞Îä• Î™®Îìà

```python
import gradio as gr

# ÏÉÅÎã¥Î¥á
def counseling_bot_chat(message, chat_history):
    if message == "":
        return "", chat_history
    else:
        completion = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "ÎãπÏã†ÏùÄ Ìó§Ïù¥ÎßàÌä∏Ïùò ÏÉÅÎã¥ÏõêÏûÖÎãàÎã§. ÎßàÌä∏ ÏÉÅÌíàÍ≥º Í¥ÄÎ†®ÎêòÏßÄ ÏïäÏùÄ ÏßàÎ¨∏ÏóêÎäî Ï†ïÏ§ëÌûà Í±∞Ï†àÌïòÏÑ∏Ïöî."},
                {"role": "user", "content": message}
            ])
        chat_history.append([message, completion.choices[0].message.content])
        return "", chat_history

# ÏÉÅÎã¥Î¥á - ÎêòÎèåÎ¶¨Í∏∞
def counseling_bot_undo(chat_history):
    if len(chat_history) > 1:
        chat_history.pop()
    return chat_history

# ÏÉÅÎã¥Î¥á - Ï¥àÍ∏∞Ìôî
def counseling_bot_reset(chat_history):
    chat_history = [[None, "ÏïàÎÖïÌïòÏÑ∏Ïöî, Ìó§Ïù¥ÎßàÌä∏ÏûÖÎãàÎã§. ÏÉÅÎã¥ÏùÑ ÎèÑÏôÄÎìúÎ¶¨Í≤†ÏäµÎãàÎã§."]]
    return chat_history

# Î≤àÏó≠Î¥á
def translate_bot(output_conditions, output_language, input_text):
    if input_text == "":
        return ""
    else:
        if output_conditions == "":
            output_conditions = ""
        else:
            output_conditions = "Î≤àÏó≠Ìï† ÎïåÏùò Ï°∞Í±¥ÏùÄ Îã§ÏùåÍ≥º Í∞ôÏäµÎãàÎã§. " + output_conditions

        completion = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "ÎãπÏã†ÏùÄ Î≤àÏó≠Í∞ÄÏûÖÎãàÎã§. ÏûÖÎ†•Ìïú Ïñ∏Ïñ¥Î•º Îã§Î•∏ ÏÑ§Î™ÖÏóÜÏù¥ Í≥ßÎ∞îÎ°ú {0}Î°ú Î≤àÏó≠Ìï¥ÏÑú ÏïåÎ†§ Ï£ºÏÑ∏Ïöî. Î≤àÏó≠Ïù¥ Î∂àÍ∞ÄÎä•Ìïú Ïñ∏Ïñ¥ÎùºÎ©¥ Î≤àÏó≠Ïù¥ Î∂àÍ∞ÄÎä•ÌïòÎã§Í≥† ÎßêÌïú ÌõÑ Í∑∏ Ïù¥Ïú†Î•º ÏÑ§Î™ÖÌï¥ Ï£ºÏÑ∏Ïöî. {1}".format(output_language, output_conditions)},
                {"role": "user", "content": input_text}
            ])

        return completion.choices[0].message.content

# ÏÜåÏÑ§Î¥á
def novel_bot(model, temperature, detail):
    completion = client.chat.completions.create(
        model=model,
        temperature=temperature,
        messages=[
            {"role": "system", "content": "ÎãπÏã†ÏùÄ ÏÜåÏÑ§Í∞ÄÏûÖÎãàÎã§. ÏöîÏ≤≠ÌïòÎäî Ï°∞Í±¥Ïóê ÎßûÏ∂∞ ÏÜåÏÑ§ÏùÑ ÏûëÏÑ±Ìï¥Ï£ºÏÑ∏Ïöî."},
            {"role": "user", "content": detail}
        ])
    return completion.choices[0].message.content
```

```python
# Tab1. ÏÉÅÎã¥Î¥á
def layout_tab_01():   
    #1
    gr.Markdown(
        value="""
        # <center>ÏÉÅÎã¥Î¥á</center>
        <center>Ìó§Ïù¥ÎßàÌä∏ ÏÉÅÎã¥Î¥áÏûÖÎãàÎã§. ÎßàÌä∏ÏóêÏÑú ÌåêÎß§ÌïòÎäî ÏÉÅÌíàÍ≥º Í¥ÄÎ†®Îêú ÏßàÎ¨∏Ïóê ÎãµÎ≥Ä
        ÎìúÎ¶ΩÎãàÎã§.</center>
        """)
    #2
    cb_chatbot = gr.Chatbot(
        value=[[None, "ÏïàÎÖïÌïòÏÑ∏Ïöî, Ìó§Ïù¥ÎßàÌä∏ÏûÖÎãàÎã§. ÏÉÅÎã¥ÏùÑ ÎèÑÏôÄÎìúÎ¶¨Í≤†ÏäµÎãàÎã§."]],
        show_label=False
    )
    with gr.Row():
        #3
        cb_user_input = gr.Text(
            lines=1,
            placeholder="ÏûÖÎ†• Ï∞Ω",
            container=False,
            scale=9
        )
        #4
        cb_send_btn = gr.Button(
            value="Î≥¥ÎÇ¥Í∏∞",
            scale=1,
            variant="primary",
            icon="https://cdn-icons-png.flaticon.com/128/12439/12439334.png"
        )
    with gr.Row():
        #5
        gr.Button(value="‚Ü© ÎêòÎèåÎ¶¨Í∏∞").click(fn=counseling_bot_undo, inputs=cb_chatbot, outputs=cb_chatbot)
        #6
        gr.Button(value="üîÑ Ï¥àÍ∏∞Ìôî").click(fn=counseling_bot_reset, inputs=cb_chatbot, outputs=cb_chatbot)
        # Î≥¥ÎÇ¥Í∏∞1
        cb_send_btn.click(fn=counseling_bot_chat, inputs=[cb_user_input, cb_chatbot], outputs=[cb_user_input, cb_chatbot])
        # Î≥¥ÎÇ¥Í∏∞2
        cb_user_input.submit(fn=counseling_bot_chat, inputs=[cb_user_input, cb_chatbot], outputs=[cb_user_input, cb_chatbot])
        
```

```python
# Tab2. Î≤àÏó≠Î¥á
def layout_tab_02():   
    #1
    gr.Markdown(
        value="""
        # <center>Î≤àÏó≠Î¥á</center>
        <center>Îã§Íµ≠Ïñ¥ Î≤àÏó≠Î¥áÏûÖÎãàÎã§.</center>
        """)
    with gr.Row():
        #2
        tb_output_conditions = gr.Text(
            label="Î≤àÏó≠ Ï°∞Í±¥",
            placeholder="ÏòàÏãú: ÏûêÏó∞Ïä§ÎüΩÍ≤å",
            lines=1,
            max_lines=3
        )
        #3
        tb_output_language = gr.Dropdown(
            label="Ï∂úÎ†• Ïñ∏Ïñ¥",
            choices=["ÌïúÍµ≠Ïñ¥", "ÏòÅÏñ¥", "ÏùºÎ≥∏Ïñ¥", "Ï§ëÍµ≠Ïñ¥"],
            value="ÌïúÍµ≠Ïñ¥",
            allow_custom_value=True,
            interactive=True
        )
    #4
    tb_submit = gr.Button(
        value="Î≤àÏó≠ÌïòÍ∏∞",
        variant="primary"
    )
    with gr.Row():
        #5
        tb_input_text = gr.Text(
            placeholder="Î≤àÏó≠Ìï† ÎÇ¥Ïö©ÏùÑ Ï†ÅÏñ¥Ï£ºÏÑ∏Ïöî.",
            lines=10,
            max_lines=20,
            show_copy_button=True,
            label=""
        )
        #6
        tb_output_text = gr.Text(
            lines=10,
            max_lines=20,
            show_copy_button=True,
            label="",
            interactive=False
        )
    # Î≤àÏó≠Î¥á ÎÇ¥Ïö© Î≥¥ÎÇ¥Í∏∞
    tb_submit.click(
        fn=translate_bot,
        inputs=[tb_output_conditions,
            tb_output_language,
            tb_input_text],
        outputs=tb_output_text
    )
```

```python
# Tab3. ÏÜåÏÑ§Î¥á
def layout_tab_03():   
    #1
    gr.Markdown(
        value="""
        # <center>ÏÜåÏÑ§Î¥á</center>
        <center>ÏÜåÏÑ§ÏùÑ ÏÉùÏÑ±Ìï¥Ï£ºÎäî Î¥áÏûÖÎãàÎã§.</center>
        """)
    with gr.Accordion(label="ÏÇ¨Ïö©Ïûê ÏÑ§Ï†ï"):
        with gr.Row():
            with gr.Column(scale=1):
                #2
                nb_model = gr.Dropdown(
                    label="Î™®Îç∏ ÏÑ†ÌÉù",
                    choices=["gpt-3.5-turbo", "gpt-3.5-turbo-16k", "gpt-4", "gpt-4-32k", "gpt-4-1106-preview"],
                    value="gpt-4-1106-preview",
                    interactive=True
                )
                #3
                nb_temperature = gr.Slider(
                    label="Ï∞ΩÏùòÏÑ±",
                    info="Ïà´ÏûêÍ∞Ä ÎÜíÏùÑ ÏàòÎ°ù Ï∞ΩÏùòÏ†Å",
                    minimum=0,
                    maximum=2,
                    step=0.1,
                    value=1,
                    interactive=True
                )
            #4
            nb_detail = gr.Text(
                container=False,
                placeholder="ÏÜåÏÑ§Ïùò ÏÑ∏Î∂ÄÏ†ÅÏù∏ ÏÑ§Ï†ïÏùÑ ÏûëÏÑ±Ìï©ÎãàÎã§.",
                lines=8,
                scale=4
            )
    #5
    nb_submit = gr.Button(
        value="ÏÉùÏÑ±ÌïòÍ∏∞",
        variant="primary"
    )
    #6
    nb_output = gr.Text(
        label="",
        placeholder="Ïù¥Í≥≥Ïóê ÏÜåÏÑ§Ïùò ÎÇ¥Ïö©Ïù¥ Ï∂úÎ†•Îê©ÎãàÎã§.",
        lines=10,
        max_lines=200,
        show_copy_button=True
    )
    # Î≥¥ÎÇ¥Í∏∞
    nb_submit.click(
        fn=novel_bot,
        inputs=[nb_model, nb_temperature, nb_detail],
        outputs=nb_output
    )
    
```

```python
import gradio as gr

#----------------------------------------------------------
# 1.Í∏∞Îä•Î™®Îìà
#----------------------------------------------------------

# ÏÉÅÎã¥Î¥á
def counseling_bot_chat(message, chat_history):
    if message == "":
        return "", chat_history
    else:
        completion = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "ÎãπÏã†ÏùÄ Ìó§Ïù¥ÎßàÌä∏Ïùò ÏÉÅÎã¥ÏõêÏûÖÎãàÎã§. ÎßàÌä∏ ÏÉÅÌíàÍ≥º Í¥ÄÎ†®ÎêòÏßÄ ÏïäÏùÄ ÏßàÎ¨∏ÏóêÎäî Ï†ïÏ§ëÌûà Í±∞Ï†àÌïòÏÑ∏Ïöî."},
                {"role": "user", "content": message}
            ])
        chat_history.append([message, completion.choices[0].message.content])
        return "", chat_history

# ÏÉÅÎã¥Î¥á - ÎêòÎèåÎ¶¨Í∏∞
def counseling_bot_undo(chat_history):
    if len(chat_history) > 1:
        chat_history.pop()
    return chat_history

# ÏÉÅÎã¥Î¥á - Ï¥àÍ∏∞Ìôî
def counseling_bot_reset(chat_history):
    chat_history = [[None, "ÏïàÎÖïÌïòÏÑ∏Ïöî, Ìó§Ïù¥ÎßàÌä∏ÏûÖÎãàÎã§. ÏÉÅÎã¥ÏùÑ ÎèÑÏôÄÎìúÎ¶¨Í≤†ÏäµÎãàÎã§."]]
    return chat_history

# Î≤àÏó≠Î¥á
def translate_bot(output_conditions, output_language, input_text):
    if input_text == "":
        return ""
    else:
        if output_conditions == "":
            output_conditions = ""
        else:
            output_conditions = "Î≤àÏó≠Ìï† ÎïåÏùò Ï°∞Í±¥ÏùÄ Îã§ÏùåÍ≥º Í∞ôÏäµÎãàÎã§. " + output_conditions

        completion = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "ÎãπÏã†ÏùÄ Î≤àÏó≠Í∞ÄÏûÖÎãàÎã§. ÏûÖÎ†•Ìïú Ïñ∏Ïñ¥Î•º Îã§Î•∏ ÏÑ§Î™ÖÏóÜÏù¥ Í≥ßÎ∞îÎ°ú {0}Î°ú Î≤àÏó≠Ìï¥ÏÑú ÏïåÎ†§ Ï£ºÏÑ∏Ïöî. Î≤àÏó≠Ïù¥ Î∂àÍ∞ÄÎä•Ìïú Ïñ∏Ïñ¥ÎùºÎ©¥ Î≤àÏó≠Ïù¥ Î∂àÍ∞ÄÎä•ÌïòÎã§Í≥† ÎßêÌïú ÌõÑ Í∑∏ Ïù¥Ïú†Î•º ÏÑ§Î™ÖÌï¥ Ï£ºÏÑ∏Ïöî. {1}".format(output_language, output_conditions)},
                {"role": "user", "content": input_text}
            ])

        return completion.choices[0].message.content

# ÏÜåÏÑ§Î¥á
def novel_bot(model, temperature, detail):
    completion = client.chat.completions.create(
        model=model,
        temperature=temperature,
        messages=[
            {"role": "system", "content": "ÎãπÏã†ÏùÄ ÏÜåÏÑ§Í∞ÄÏûÖÎãàÎã§. ÏöîÏ≤≠ÌïòÎäî Ï°∞Í±¥Ïóê ÎßûÏ∂∞ ÏÜåÏÑ§ÏùÑ ÏûëÏÑ±Ìï¥Ï£ºÏÑ∏Ïöî."},
            {"role": "user", "content": detail}
        ])
    return completion.choices[0].message.content


#----------------------------------------------------------
# 2. Î†àÏù¥ÏïÑÏõÉ Î™®Îìà
#----------------------------------------------------------

# Tab1. ÏÉÅÎã¥Î¥á
def layout_tab_01():   
    #1
    gr.Markdown(
        value="""
        # <center>ÏÉÅÎã¥Î¥á</center>
        <center>Ìó§Ïù¥ÎßàÌä∏ ÏÉÅÎã¥Î¥áÏûÖÎãàÎã§. ÎßàÌä∏ÏóêÏÑú ÌåêÎß§ÌïòÎäî ÏÉÅÌíàÍ≥º Í¥ÄÎ†®Îêú ÏßàÎ¨∏Ïóê ÎãµÎ≥Ä
        ÎìúÎ¶ΩÎãàÎã§.</center>
        """)
    #2
    cb_chatbot = gr.Chatbot(
        value=[[None, "ÏïàÎÖïÌïòÏÑ∏Ïöî, Ìó§Ïù¥ÎßàÌä∏ÏûÖÎãàÎã§. ÏÉÅÎã¥ÏùÑ ÎèÑÏôÄÎìúÎ¶¨Í≤†ÏäµÎãàÎã§."]],
        show_label=False
    )
    with gr.Row():
        #3
        cb_user_input = gr.Text(
            lines=1,
            placeholder="ÏûÖÎ†• Ï∞Ω",
            container=False,
            scale=9
        )
        #4
        cb_send_btn = gr.Button(
            value="Î≥¥ÎÇ¥Í∏∞",
            scale=1,
            variant="primary",
            icon="https://cdn-icons-png.flaticon.com/128/12439/12439334.png"
        )
    with gr.Row():
        #5
        gr.Button(value="‚Ü© ÎêòÎèåÎ¶¨Í∏∞").click(fn=counseling_bot_undo, inputs=cb_chatbot, outputs=cb_chatbot)
        #6
        gr.Button(value="üîÑ Ï¥àÍ∏∞Ìôî").click(fn=counseling_bot_reset, inputs=cb_chatbot, outputs=cb_chatbot)
        # Î≥¥ÎÇ¥Í∏∞1
        cb_send_btn.click(fn=counseling_bot_chat, inputs=[cb_user_input, cb_chatbot], outputs=[cb_user_input, cb_chatbot])
        # Î≥¥ÎÇ¥Í∏∞2
        cb_user_input.submit(fn=counseling_bot_chat, inputs=[cb_user_input, cb_chatbot], outputs=[cb_user_input, cb_chatbot])
        

# Tab2. Î≤àÏó≠Î¥á
def layout_tab_02():   
    #1
    gr.Markdown(
        value="""
        # <center>Î≤àÏó≠Î¥á</center>
        <center>Îã§Íµ≠Ïñ¥ Î≤àÏó≠Î¥áÏûÖÎãàÎã§.</center>
        """)
    with gr.Row():
        #2
        tb_output_conditions = gr.Text(
            label="Î≤àÏó≠ Ï°∞Í±¥",
            placeholder="ÏòàÏãú: ÏûêÏó∞Ïä§ÎüΩÍ≤å",
            lines=1,
            max_lines=3
        )
        #3
        tb_output_language = gr.Dropdown(
            label="Ï∂úÎ†• Ïñ∏Ïñ¥",
            choices=["ÌïúÍµ≠Ïñ¥", "ÏòÅÏñ¥", "ÏùºÎ≥∏Ïñ¥", "Ï§ëÍµ≠Ïñ¥"],
            value="ÌïúÍµ≠Ïñ¥",
            allow_custom_value=True,
            interactive=True
        )
    #4
    tb_submit = gr.Button(
        value="Î≤àÏó≠ÌïòÍ∏∞",
        variant="primary"
    )
    with gr.Row():
        #5
        tb_input_text = gr.Text(
            placeholder="Î≤àÏó≠Ìï† ÎÇ¥Ïö©ÏùÑ Ï†ÅÏñ¥Ï£ºÏÑ∏Ïöî.",
            lines=10,
            max_lines=20,
            show_copy_button=True,
            label=""
        )
        #6
        tb_output_text = gr.Text(
            lines=10,
            max_lines=20,
            show_copy_button=True,
            label="",
            interactive=False
        )
    # Î≤àÏó≠Î¥á ÎÇ¥Ïö© Î≥¥ÎÇ¥Í∏∞
    tb_submit.click(
        fn=translate_bot,
        inputs=[tb_output_conditions,
            tb_output_language,
            tb_input_text],
        outputs=tb_output_text
    )

# Tab3. ÏÜåÏÑ§Î¥á
def layout_tab_03():   
    #1
    gr.Markdown(
        value="""
        # <center>ÏÜåÏÑ§Î¥á</center>
        <center>ÏÜåÏÑ§ÏùÑ ÏÉùÏÑ±Ìï¥Ï£ºÎäî Î¥áÏûÖÎãàÎã§.</center>
        """)
    with gr.Accordion(label="ÏÇ¨Ïö©Ïûê ÏÑ§Ï†ï"):
        with gr.Row():
            with gr.Column(scale=1):
                #2
                nb_model = gr.Dropdown(
                    label="Î™®Îç∏ ÏÑ†ÌÉù",
                    choices=["gpt-3.5-turbo", "gpt-3.5-turbo-16k", "gpt-4", "gpt-4-32k", "gpt-4-1106-preview"],
                    value="gpt-4-1106-preview",
                    interactive=True
                )
                #3
                nb_temperature = gr.Slider(
                    label="Ï∞ΩÏùòÏÑ±",
                    info="Ïà´ÏûêÍ∞Ä ÎÜíÏùÑ ÏàòÎ°ù Ï∞ΩÏùòÏ†Å",
                    minimum=0,
                    maximum=2,
                    step=0.1,
                    value=1,
                    interactive=True
                )
            #4
            nb_detail = gr.Text(
                container=False,
                placeholder="ÏÜåÏÑ§Ïùò ÏÑ∏Î∂ÄÏ†ÅÏù∏ ÏÑ§Ï†ïÏùÑ ÏûëÏÑ±Ìï©ÎãàÎã§.",
                lines=8,
                scale=4
            )
    #5
    nb_submit = gr.Button(
        value="ÏÉùÏÑ±ÌïòÍ∏∞",
        variant="primary"
    )
    #6
    nb_output = gr.Text(
        label="",
        placeholder="Ïù¥Í≥≥Ïóê ÏÜåÏÑ§Ïùò ÎÇ¥Ïö©Ïù¥ Ï∂úÎ†•Îê©ÎãàÎã§.",
        lines=10,
        max_lines=200,
        show_copy_button=True
    )
    # Î≥¥ÎÇ¥Í∏∞
    nb_submit.click(
        fn=novel_bot,
        inputs=[nb_model, nb_temperature, nb_detail],
        outputs=nb_output
    )



#----------------------------------------------------------
# 3. Ï±óÎ¥á Ïã§Ìñâ
#----------------------------------------------------------

def main():
    with gr.Blocks(theme=gr.themes.Default()) as app:
        with gr.Tab("ÏÉÅÎã¥Î¥á"):
            layout_tab_01()   

        with gr.Tab("Î≤àÏó≠Î¥á"):
            layout_tab_02()

        with gr.Tab("ÏÜåÏÑ§Î¥á"):
            layout_tab_03()

    # Ïã§Ìñâ
    app.launch()


if __name__ == "__main__":
    main()

```
